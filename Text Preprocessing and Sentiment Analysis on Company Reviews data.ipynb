The dataset contains Text reviews. The aim of this assignment is to classsify the review texts into two categories. 

Attribute information: 
 
Dataset has two columns:
Column2: Review Text
Column1: Two Classes. Positive and negative.

 <font color='brown'> The dataset contains the business and financial aspects of several companies, providing information on their performance, plans, and market activities. There are 1967 entries and each one is  classified into positive or negative based on their reviews.</font>

```python
df['Sentiment']=df['Sentiment'].replace({'negative':'0','positive':'1'})
df.head(5)
```

```python
from bs4 import BeautifulSoup
import re
```

```python
def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()
```

```python
def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
df['Review'] = df['Review'].apply(remove_html_tags)
df['Review'] = df['Review'].apply(remove_special_chars)
df.head(5)
```

 <font color='brown'> All special characters, numbers and html tags are removed from the dataset.</font>

```python
df['Review'] = df['Review'].str.lower()
df.head(5)
```

 <font color='brown'> Entire text data is converted to lowercase.</font>

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stp_wds = stopwords.words('english')
stp_wds.remove('not')
print(stp_wds)
```

```python
def remove_stopwords(text):
    words = text.split()  # Split the text into words
    filtered_words = [i for i in words if i not in set(stp_wds)]
    return ' '.join(filtered_words)  # Join the filtered words back into a sentence
```

```python
df['Review'] = df['Review'].apply(remove_stopwords)
df.head(5)
```

 <font color='brown'> Stop words like articles, pronouns, common verb, prepositions, conjunctions are removed.</font>

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem_text(text):
    words = text.split()  # Tokenize the text into words
    stemmed_words = [ps.stem(i) for i in words]  # Stem each word
    return ' '.join(stemmed_words)  # Join the stemmed words back into a sentence
```

```python
df['Review'] = df['Review'].apply(stem_text)
df.head(10)
```

 <font color='brown'> Stemming is the process of removing inflections from the word and getting the base or root word.</font>

<font color='brown'>"Bag of Words" (BoW) is a common technique in natural language processing (NLP) and text analysis. It converts a collection of text documents into numerical features that can be used to give inputs into an ML model.</font>

```python
corpus = []
for i in range(0, len(df)):
    corpus.append(df['Review'][i])
corpus    
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
X = cv.fit_transform(corpus).toarray()
X
```

```python
samp=df.iloc[:]
samp.head(5)
```

```python
Y= df.iloc[:,0]
Y.head(5)
```

```python
Y=df.iloc[:,0].values
Y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 40)
```

4.3 - Define the Classification model (any one like NB, SVM or Random forest) train the model

```python
# Training model using Naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(X_train, y_train)
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion_matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```

```python
print("Accuracy_score:",accuracy_score(y_test, y_pred))
```

 <font color='brown'>The accuracy score of the ML model constructed by using Multinomial Naive Bayes classifier and CountVectorizer is 0.84 </font>

```python
test_review1 = "In the last quarter of 2010, Espresso financials witnessed a remarkable performance."
tst_review_bow1 = cv.transform([test_review1])
tst_pred1 = classifier.predict(tst_review_bow1)

if tst_pred1 == ['1']:
    print("Predicted class of test review is :",tst_pred1, "--> POSITIVE")
else:
    print("Predicted class of test review is :",tst_pred1, "--> NEGATIVE")
```

```python
test_review2 = "Brownwich Newyork recently laid off several employees at its 4 sites in the city"
tst_review_bow2 = cv.transform([test_review2])
tst_pred2 = classifier.predict(tst_review_bow2)

if tst_pred2 == ['1']:
    print("Predicted class of test review is :",tst_pred2, "--> POSITIVE")
else:
    print("Predicted class of test review is :",tst_pred2, "--> NEGATIVE")
```

 <font color='green'> The test-review results are predicted correctly as expected. The output of review1 is  "positive" as the tone of review is clearly talking about the good performance of the company. The output of review2 is "negative" as its says company has laid down employees which might be due to poor profits in that fiscal year.</font>

```python
# Creating the TFIDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=4000)
X_tf = tv.fit_transform(corpus).toarray()
X_tf
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
Xtf_train, Xtf_test, ytf_train, ytf_test = train_test_split(X_tf, Y, test_size = 0.30, random_state = 40)
```

```python
from sklearn.naive_bayes import MultinomialNB
tf_classifier = MultinomialNB().fit(Xtf_train, ytf_train)
ytf_pred=tf_classifier.predict(Xtf_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_tf = confusion_matrix(ytf_test, ytf_pred)
print("Confusion_matrix:\n", cm_tf)
```

```python
from sklearn.metrics import classification_report
report_tf = classification_report(ytf_test, ytf_pred)
print(report_tf)
```

```python
print("Accuracy_score:",accuracy_score(ytf_test, ytf_pred))
```

 <font color='brown'>The accuracy score of the ML model constructed by using Multinomial Naive Bayes classifier and TF-IDF Vectorizer is 0.765 </font>
